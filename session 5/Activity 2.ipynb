```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def Football_3D(v0, theta, dt, g):
    x_array, y_array, z_array = [x0], [y0], [z0]
    t=0
    horizontal_distance = (x_target ** 2 + y_target ** 2) ** 0.5
    horizontal_speed = v0 * np.cos(theta)
    x_speed = horizontal_speed * x_target / horizontal_distance
    y_speed = horizontal_speed * y_target / horizontal_distance
    z_speed = v0 * np.sin(theta)
    while t < horizontal_distance / horizontal_speed:
        t+=dt
        x_array.append(x_array[-1] + dt * x_speed)
        y_array.append(y_array[-1] + dt * y_speed)
        z_array.append(z_array[-1] + dt * z_speed)
        z_speed += dt * g
    return x_array, y_array, z_array

x0 = 0
y0 = 0
z0 = 0
x_target = 11
y_target = -3.5
z_target = 2.4

theta = np.pi/3
dt = 0.001
g = -9.8

for v in range(10,5000):
    v0=v/100
    x,y,z=Football_3D(v0, theta, dt, g)
    if abs(z[-1]-z_target)<0.01:
        print("V0 is approximately to be:",v0)
        break

x,y,z=Football_3D(12.18, theta, dt, g)

gate_y=np.linspace(-3.6,3.6,100)
gate_z=np.linspace(0,2.6,100)
gate_x=np.linspace(11,11,100)
gate_y, gate_z = np.meshgrid(gate_y, gate_z)

plt3d = plt.figure().gca(projection='3d')
plt3d.plot3D(x,y,z)
plt3d.plot_surface(gate_x,gate_y,gate_z)
plt.show()
```
