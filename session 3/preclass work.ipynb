```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
```

## General Solution
The Laplace Equation on a circle of radius $r_0$ with boundary conditions $u(r,\theta)=f(\theta)$ has the general solution of:
$$\sum_{i=1}^{n} a_n r_0^n cos(n\theta) + \sum_{i=1}^{n} b_n r_0^n sin(n\theta)$$

We use the boundary condition $u(r,\theta)=f(\theta)$ to narrow down the general solution to one solution.

Next, we use the solution $u(r, \theta)$ to calculate the point at each grid.

## Preclass work
Apply your algorithm when $r_0=3$, $f(\theta)=2sin(2theta)$

Solution:
$$ n = 2 and a_n = 0$$
$$ 2sin(2\theta)=b_2 3^2 sin (2\theta)$$
$$ b_2 = \frac{2 sin (2\theta)} {3^2 sin (2\theta)}$$
$$ = \frac{2}{9}$$

$$ u(r, \theta) = \frac{2}{9}r^2sin(2\theta)$$


```python
r = np.linspace(0,3,100)
theta = np.linspace(0, 2*np.pi, 100)
radius_matrix, theta_matrix = np.meshgrid(r,theta)
# X = radius_matrix * np.cos(theta_matrix)
# Y = radius_matrix * np.sin(theta_matrix)
Z = np.zeros((len(radius_matrix), len(theta_matrix)))

for ir, r_i in enumerate(r.tolist()):
    for it, theta_i in enumerate(theta.tolist()):
        # calculate u = 2/9 * r^2 sin(2 theta)
        u_i = 2/9 * r_i**2 * np.sin(2*theta_i)
        Z[ir][it] = u_i
```

```python
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
C = ax.contourf(theta_matrix, radius_matrix, Z, 50,
                 alpha = 1, origin='lower', extend='both')
cbar = fig.colorbar(C)
```
