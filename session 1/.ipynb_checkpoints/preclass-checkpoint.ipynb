```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

## Part 1

```python
url = "https://gist.githubusercontent.com/raquelhr/d1324510056353feeccf111d6b186a0d/raw/7b3bccc7917f3baa7ec1d919195d120083ee75e9/proctatinium_data.csv"
data = pd.read_csv(url)
t_values = np.array(data["time"], dtype=float) 
count_values = np.array(data["count_rate"], dtype=float)
```

## Part 2

```python
# plot 
plt.scatter(t_values, count_values)
plt.xlabel("Time")
plt.ylabel("Count")
plt.show()
```

I used scatter plot here because experimental data uses discrete time intervals, as opposed to a continuous time interval. We are only able to measure the counts are particular times.

## Part 4

Given the initial count rate ($N_0$), we can use the derivation in part 3 to compute the theoretical amount of protactinium available for every t_value in the dataset data:

```python
N_0 = data['count_rate'][0]
lambd = np.log(2)/(1.17*60) # 1.17 minutes is the half-life
predicted = N_0 * np.exp(-lambd * t_values)
```

```python
plt.scatter(t_values, count_values, label="experimental")
plt.scatter(t_values, predicted, label="predicted")
plt.xlabel("Time")
plt.ylabel("Count")
plt.legend()
plt.show()
```

We can also make a continuous graph using our equation:

```python
t = np.linspace(0, 410, 1000)
predicted = N_0 * np.exp(-lambd * t)
```

```python
plt.plot(t, predicted, label="predicted")
plt.xlabel("Time")
plt.ylabel("Count")
plt.legend()
plt.show()
```

## Part 5
We can see that the experimental data started off quite accurate to our theoretical predictions. As time goes by, however, the experimental data starts consistently overestimating the counts, compared to our theoretical prediction.
