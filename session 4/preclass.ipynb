## Update Functions
$$ \dot x(t)= \dot(t_0) + [-\omega^2 x](t-t_0)$$
$$ x(t) = x(t_0) + \dot x(t) (t-t_0)$$
where $\omega^2=\frac{k}{m}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def springSystemSimulation(x_0, xdot_0, k, m, tau, num, EC_method=False):
    # vectors
    time_vec = np.zeros(num)
    x_vec = np.zeros(num)
    xdot_vec = np.zeros(num)
    PE_vec = np.zeros(num)
    KE_vec = np.zeros(num)
    
    # initial conditions
    x = x_0
    xdot = xdot_0
    omega2 = k/m
    time_step = 0
    
    # iterating
    for i in range(num):
        xdot_old = xdot
        x_old = x
        
        # updates
        xdot = xdot_old + (-omega2*x_old) * tau
        if EC_method:
            x = x_old + xdot * tau # Euler-Cromer method
        else:
            x = x_old + xdot_old * tau
        # save values in vectors
        time_vec[i] = tau*i
        x_vec[i] = x
        xdot_vec[i] = xdot
        KE_vec[i] = (1/2)*m*xdot**2
        PE_vec[i] = (1/2) * k * x**2
    TE_vec = np.add(KE_vec,PE_vec)
    

    plt.figure(0)
    plt.plot(time_vec,x_vec)
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement')
    plt.savefig('plot1.png', bbox_inches='tight')

    plt.figure(1)
    plt.plot(time_vec,KE_vec,label='Kinetic Energy')
    plt.plot(time_vec,PE_vec,label='Potential Energy')
    plt.plot(time_vec,TE_vec,label='Total Energy')
    plt.legend(loc='upper left')
    plt.xlabel('Time (s)')
    plt.ylabel('Energy (J)')
    plt.savefig('plot2.png', bbox_inches='tight')

    plt.figure(2)
    plt.plot(x_vec,xdot_vec)
    plt.xlabel('Displacement')
    plt.ylabel('Velocity')
    plt.savefig('plot3.png', bbox_inches='tight')

    plt.show()
```

```python
springSystemSimulation(0.5, 0, 1, 5, 1, 100)
```

```python
springSystemSimulation(0.5, 0, 1, 5, 1, 100, EC_method=True)
```
