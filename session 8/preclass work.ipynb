You can modify your code for Assignment1 Q2 to solve this question. Assume there is a 25 by 25 square room, and a 3 by 3-grid 100C griller which is diffusing heat in the center of the space. The wall is well-isolated. Under a certain diffusion coefficient , simulate the temperature distribution in the room. Try to adjust the value of the diffusion coefficient and observe how it changes the rate of heat diffusion.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class roomTempSimulation:
    def __init__(self, diffusion_coef=0.01, room_length=25, deltax=1, deltaT=1, temp_heater=100, init_temp=0,
                 heater_center=(13,13), heater_length=3, wall_temp=20, max_iter=400, show_every=100):
        self.diffusion = diffusion_coef
        self.room_length = room_length
        self.deltax = deltax
        self.deltaT = deltaT
        self.temp_heater=100
        self.init_temp=init_temp
        self.heater_center = heater_center
        self.heater_length = heater_length
        self.max_iter = max_iter
        self.wall_temp = wall_temp
        self.show_every=show_every
        # macrovariables
        self.timestep = 0
        self.time = []
        self.mean_temp = []
        
        # variables for the board, they will be defined in the selfsetUp()        
        self.X = None
        self.Y = None
        self.T = None
        self.heater_coords = []
        
        
    def setUp(self):
        '''
        This function sets up the main board to its initial condition
        '''
        # set up board 
        self.X, self.Y = np.meshgrid(np.arange(0, self.room_length), np.arange(0, self.room_length))
        self.T = np.empty((self.room_length, self.room_length))
        self.T.fill(self.init_temp)
        
        # set up temperature for the special coordinates
        for i in range(self.heater_length):
            heater_x = self.heater_center[0]+i-self.heater_length//2
            for j in range(self.heater_length):
                heater_y = self.heater_center[1]+j-self.heater_length//2
                self.T[heater_x, heater_y] = self.temp_heater
                self.heater_coords.append((heater_x, heater_y))
        # record macroscopic variables at time=0
        self.time.append(self.timestep)
        self.mean_temp.append(self.T.mean())
        
        # wall temps 
        for i in range(self.room_length):
            self.T[i, 0] = self.wall_temp
            self.T[i, self.room_length-1] = self.wall_temp
            self.T[0, i] = self.wall_temp
            self.T[self.room_length-1, i] = self.wall_temp
        
        
    def step(self):
        old_T =self.T.copy()
        self.timestep +=1
        # first update non-special coordinates
        for i in range(1, self.room_length-1, self.deltax):
            for j in range(1, self.room_length-1, self.deltax):
                if (i,j) not in self.heater_coords:
                    # normal updates
                    delta = self.diffusion * ((old_T[i-1,j]+old_T[i+1,j] - 2*old_T[i,j])+(old_T[i,j-1]+old_T[i,j+1] - 2*old_T[i,j]))/self.deltax**2
                    self.T[i,j] = old_T[i,j] + delta * self.deltaT
        
        self.time.append(self.timestep)
        self.mean_temp.append(self.T.mean())
        
    def main(self):
        '''
        This is the main function that runs the simulation
        It runs the simulation according to max_iteration
        '''
        self.setUp()
        colorinterpolation = 50
        colourMap = plt.cm.jet 
        for iteration in range(0, self.max_iter):
            if iteration%self.show_every == 0:
                #print(self.T)
                plt.title(f"Contour of Temperature {iteration} steps")
                plt.contourf(self.X, self.Y, self.T, colorinterpolation, cmap=colourMap)
                plt.colorbar()
                plt.show()
            self.step()
        # Last step
        plt.title(f"Contour of Temperature {iteration} steps")
        plt.contourf(self.X, self.Y, self.T, colorinterpolation, cmap=colourMap)
        plt.colorbar()
        plt.show()
        plt.title("Temperature Evolution over time")
        plt.plot(self.time, self.mean_temp)
        plt.ylabel("Mean Temperature of the room")
        plt.xlabel("Timesteps")
        plt.show()
```

```python
sim=roomTempSimulation(max_iter=8000, show_every=2000)
sim.main()
```

```python
sim=roomTempSimulation(diffusion_coef=0.15, max_iter=8000, show_every=2000)
sim.main()
```

Higher diffusion coefficients results in much faster convergence in our simulations
